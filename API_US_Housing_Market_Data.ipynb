#### <font color="purple">1. National Region</font> üèò

##### <font color="orange">Standard Query</font> ‚ú®

```python
url = "https://us-housing-market-data.p.rapidapi.com/getNational"

headers = {
	"X-RapidAPI-Key": rapid_api_key,
	"X-RapidAPI-Host": "us-housing-market-data.p.rapidapi.com"
}

response = requests.request("GET", url, headers=headers)

print(response.status_code)
```

### <font color="green">Section #2 - Loops</font> üèô
Get data for multiple regions

```python
# get all neighborhoods in a single city
city = 'Tampa, FL'
df_neighborhood_fltr = df_neighborhood_ids.loc[
    df_neighborhood_ids['region'].str.contains(city)]
neighborhood_ids_list = df_neighborhood_fltr['region_id'].tolist()
print('Num of neighborhoods:', len(neighborhood_ids_list))
print('Sample neighborhoods:', df_neighborhood_fltr['region'].tolist()[:5])
```

```python
subset_neighborhoods_list = ['Tampa, FL - Lowry Park North',
 'Tampa, FL - Lowry Park Central',
 'Tampa, FL - Historic Ybor City',
 'Tampa, FL - East Ybor Historic',
 'Tampa, FL - V. M. Ybor',
 'Tampa, FL - North Ybor',
 'Tampa, FL - Ybor Heights',
 'Tampa, FL - Ybor City Historic District',
 'Tampa, FL - Old Seminole Heights'
 'Tampa, FL - South Seminole Heights', 
 'Tampa, FL - Southeast Seminole Heights',
 'Tampa, FL - East Seminole Heights',
 'Tampa, FL - Seminole Heights Historic District',
 'Tampa, FL - Seminole Heights',
 'Tampa, FL - Riverside Heights',
 'Tampa, FL - West Riverside Heights',
 'Tampa, FL - Eastern Heights',
 'Tampa, FL - Tampa Heights',
 'Tampa, FL - Tampa Heights Historic District'
 'Tampa, FL - Hyde Park Historic District',
 'Tampa, FL - Hyde Park North',
 'Tampa, FL - Hyde Park',
 'Tampa, FL - North Hyde Park',
 'Tampa, FL - Historic Hyde Park North']
subset_ids_list = df_neighborhood_fltr\
  .loc[df_neighborhood_fltr['region'].isin(subset_neighborhoods_list)]\
  ['region_id'].tolist()
print('Num of neighborhoods in subset: {}'.format(str(len(subset_ids_list))))
```

```python
# get neighborhood data
df_list = [] # empty list

# iterate through each neighborhood id in the list
for n_id in subset_ids_list:
  url = "https://us-housing-market-data.p.rapidapi.com/getNeighborhood"

  querystring = {"neighborhood_id": n_id}

  headers = {
    "X-RapidAPI-Key": rapid_api_key,
    "X-RapidAPI-Host": "us-housing-market-data.p.rapidapi.com"
  }

  # make request
  response = requests.request("GET", url, headers=headers, params=querystring)
  
  # convert to dataframe
  _df = pd.DataFrame.from_dict(response.json(), orient='index') 
  df_list.append(_df) # append dataframe to list
  time.sleep(0.2) # pause (Pro tier = 5 requests per second)
```

```python
df_merge = pd.concat(df_list)
print('Num of rows: {}'.format(len(df_merge)))
print('Num of columns: {}'.format(len(df_merge.columns)))
df_merge.head()
```

### <font color="green">Section #3 - Visualization</font> üìà
Visualize housing and economic data trends over time

```python
max_period = df_merge['period_begin'].max()
print('Max period: {}'.format(max_period))
```

```python
# filter property type = single family
df_sf = df_merge.loc[
    (df_merge['property_type'] == 'Single Family Residential')]\
    .sort_values(by="period_begin")
```

##### <font color="orange">‚ùì Which neighborhoods are most expensive ‚ùì</font>

```python
df_sf.loc[(df_sf['period_begin'] == max_period)]\
  .sort_values(by=['median_sale_price'], ascending=False)\
  [['region', 'median_sale_price']].head()
```

```python
df_plot = df_sf.loc[(df_sf['period_begin'] == max_period)]
fig = px.box(df_plot, y="median_sale_price", title="Median Sale Price")
fig.show()
```

##### <font color="orange">‚ùì Which neighborhoods appreciated the most ‚ùì</font>

```python
df_sf.loc[(df_sf['period_begin'] == max_period)]\
  .sort_values(by=['median_sale_price_yoy'], ascending=False)\
  [['region', 'median_sale_price_yoy']].head()
```

```python
df_plot = df_sf.loc[(df_sf['period_begin'] == max_period)]\
  .sort_values(by=['median_sale_price_yoy'], ascending=False)
fig = px.bar(df_plot, x='region', y='median_sale_price_yoy', title='Median Sale Price YoY')
fig.show()
```

##### <font color="orange">‚ùì Historical ‚ùì</font>

```python
fig = px.line(df_sf, x="period_begin", y="median_sale_price", 
              color="region", title='Median Sale Price')
fig.show()
```

```python
fig = px.line(df_sf, x="period_begin", y="median_ppsf", 
              color="region", title='Median PPSF')
fig.show()
```
